```python
# importing dependencies
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import re

import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading the dataset
df = pd.read_csv(r'SanFrancisco Crime Incidents.csv')
# Check the shape for the dataset
df.shape
```

```python
df.head()
```

```python
# Check for null values in the dataset
df.isnull().sum()
```

```python
# Let's fill the missing value in the PdDistrict Column with mode of the column
df['PdDistrict'].fillna(df['PdDistrict'].mode()[0], inplace = True)
df['PdDistrict'].isnull().sum()
```

```python
# Dropping the unnecessary columns 
df.drop(labels=['IncidntNum','PdId'], axis = 1, inplace = True)
df.head()
```

```python
# Rename columns
df.rename(columns={'Descript': 'Description', 'X':'Longitude', 'Y':'Latitude'}, inplace=True)
df.columns
```

```python
# Modyfying the date format as per our convenience
df['Date'] = df['Date'].str.extract('([0-9]{2}\/[0-9]{2}\/[0-9]{4})')
date = df['Date'].str.split('/')
date_list = []
for i in date:
    date = '{}-{}-{}'.format(i[1], i[0], i[2])
    date_list.append(date)
df['Date'] = date_list
```

```python
df.head()
```

```python
# DataFrame to Csv file
df.to_csv('SanFrancisco Crime Incidents_Clean.csv', index = False)
```
